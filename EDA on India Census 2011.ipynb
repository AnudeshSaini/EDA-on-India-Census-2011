## Population Stats
1) What are the population of states ?

2) Which are the top 10 populated states in India ?

3) Which are the least 10 populated states in India ?

 * Uttar pardesh is most populated in India with 199812341 population.
 * Lakshdweep is least populated in India with 64473 population.

```python
# Visualization of Population of states
pop = census.groupby("State_name")[["Population"]].sum().sort_values(by=["Population"],ascending=False)
plt.plot(pop.index , pop.Population,c='r', marker ='o' )
plt.bar(pop.index , pop.Population)
plt.xlabel("State name",fontsize=20)
plt.xticks(rotation=70)
plt.ylabel("Population (10 Crores)",fontsize=20)
plt.title("Population of States", fontsize=20, color="Black")
matplotlib.rcParams["figure.figsize"] = (20,10)
plt.grid(True)
```

```python
# Visualization of Top 10 populated states
pop_10 = census.groupby('State_name')[['Population']].sum().sort_values('Population', ascending = False)[:10]
plt.figure(figsize=(20,10))                                                     
plt.plot(pop_10.index, pop_10.Population, c='r', marker ='o')
plt.bar(pop_10.index, pop_10.Population)
plt.title('Top 10 most populated states/Uts', c='r', fontsize = 20)
plt.xlabel('States/Uts', c='r', fontsize = 20)
plt.ylabel('Population(10 crore)', c='r', fontsize = 20);
plt.grid(True)
plt.show()
```

```python
#Visualization of 10 least populated states
low_10 = census.groupby("State_name")[["Population"]].sum().sort_values(by="Population",ascending= True)[:10]
plt.bar(low_10.index, low_10.Population, width=0.8 , align="center")
plt.plot(low_10.index, low_10.Population, c='r', marker ='o')
plt.title('Least 10 populated states/Uts', fontsize=20, c='r')
plt.xlabel('States/Uts', fontsize = 20, c='r')
plt.ylabel('Population', fontsize = 20, c='r')
plt.xticks(fontsize=12, rotation= 0)
plt.yticks(fontsize=12)
plt.grid(True)
matplotlib.rcParams["figure.figsize"]=(20,10)
plt.show()
```

## Male & Female population

1) What is the Male & Female Percentage in Country ?

* Percentage of Male in India = 51.473567837513215 % 

* Percentage of Female in India = 48.526432162486785 %

```python
# Male & Female Percentage in Country
total_pop = census.Population.sum()
male_pop = census.Male.sum()
female_pop = census.Female.sum()
male_percent = (male_pop/total_pop)*100
female_percent = (female_pop/total_pop)*100
print( 'Percentage of Male in India = {}'.format(male_percent),'%')
print( 'Percentage of Female in India = {}'.format(female_percent),'%')
```

##  Sex Ratio
1) Calculate & create a new column for "sex_ratio"

2) Which are the best  10 States by "sex ratio" ?

3) Which are the worst  10 States by "sex ratio" ?
 * Kerala have best "sex_ratio" in India
 
 * Dadra and Nagar Haveli have Worst "sex_ratio" in India

```python
census["sex_ratio"] = (census["Female"]/census["Male"])*1000
```

```python
sex_ratio = census.groupby("State_name")[["sex_ratio"]].mean().sort_values(by= ["sex_ratio"], ascending=False).head(10)
plt.bar(sex_ratio.index , sex_ratio["sex_ratio"] , color="Green" , width=0.5)
plt.plot(sex_ratio.index , sex_ratio["sex_ratio"] , color="red",marker= ".", markersize=12)
matplotlib.rcParams["figure.figsize"]= (20,6)
plt.grid(True)
plt.show()
```

```python
sex_ratio = census.groupby("State_name")[["sex_ratio"]].mean().sort_values(by= ["sex_ratio"], ascending=True).head(10)
plt.bar(sex_ratio.index , sex_ratio["sex_ratio"] , color="Green" , width=0.5)
plt.plot(sex_ratio.index , sex_ratio["sex_ratio"] , color="red",marker= ".", markersize=12)
matplotlib.rcParams["figure.figsize"]= (20,6)
plt.grid(True)
plt.show()
```

## Religion Stats
1) Whats is the weight of different religions in India ?

* Out of all regions hindus have highest population with 80.8% 
* Out of all regions Buddhist have least population with 0.7% 

```python
hindus = census["Hindus"].sum()
muslims =  census["Muslims"].sum()
christians = census["Christians"].sum()
sikhs = census["Sikhs"].sum()
buddhists = census["Buddhists"].sum()
religions = [hindus, muslims,christians, sikhs ,buddhists]
name = ["Hindus", "Muslims", "Christians", "Sikhs","Buddhists"]
plt.pie(religions, labels= name ,colors = ["orange" , "green", "red", "pink", "brown"], startangle=90, autopct = "%0.1f%%" , explode=[0.1,0.1,0.2,0.3,0.4], radius= 1.2 )
plt.legend (name)
plt.show()
```
